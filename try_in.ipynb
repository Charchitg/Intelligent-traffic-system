```python
import numpy as np
import cv2

image_11 = cv2.imread('pic-1.jpg')
image_12 = cv2.imread('pic-2.jpg')



height_11, width_11 = image_11.shape[:2]
height_12, width_12 = image_12.shape[:2]

#  get the starting pixel coordiantes (top  left of cropping rectangle)
start_row_11, start_col_11 = int(height_11 * .05), int(width_11 * .25)

#  get the ending pixel coordinates (bottom right)
end_row_11, end_col_11 = int(height_11 * .80), int(width_11 * .48)

# Simply use indexing to crop out the rectangle we desire
image_13 = image_11[start_row_11:end_row_11 , start_col_11:end_col_11]

#  get the starting pixel coordiantes (top  left of cropping rectangle)
start_row_12,start_col_12 = int(height_12 * .05), int(width_12 * .25)

#  get the ending pixel coordinates (bottom right)
end_row_12, end_col_12 = int(height_12 * .80), int(width_12 * .48)

# Simply use indexing to crop out the rectangle we desire

image_14 = image_12[start_row_12:end_row_12 , start_col_12:end_col_12]

# taking absoloute difference
diff1 = cv2.absdiff(image_13.copy(),image_14.copy())

# Gray_scale image
gray1 = cv2.cvtColor(diff1 , cv2.COLOR_BGR2GRAY)


# blurring the image
blur1 = cv2.GaussianBlur(gray1 , (5,5) , 0)


#thresholding
ret,thresh1 = cv2.threshold(blur1 , 20 , 255 , cv2.THRESH_BINARY)


# dilating the image
dilated1 = cv2.dilate(thresh1 , None , iterations = 1)


contours1,hierarchy = cv2.findContours(dilated1.copy() , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
#counting valid contours                    
valid_cntrs1 = []
for contour in contours1:
    (x,y,w,h) = cv2.boundingRect(contour)
    if cv2.contourArea(contour) < 300  :
        continue               
                    
    cv2.rectangle(image_14 , (x,y) , (x+w , y+h) , (0 , 255 , 0) , 2) 
    valid_cntrs1.append(contour)
    
                   

                    
                    
feed1 = image_14.copy()
                    
cv2.putText(feed1, "Cars Detected : {}".format(str(len(valid_cntrs1))) , (10,20) , cv2.FONT_HERSHEY_SIMPLEX , 1 , (255 , 255 , 255) , 1)
                    
cv2.imshow("output1" , feed1)




cv2.waitKey(0)
cv2.destroyAllWindows()
```
